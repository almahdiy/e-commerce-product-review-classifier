```python
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(df['Review'])):
    words = nltk.word_tokenize(df['Review'][i])              # Tokenize into words.
    words = [lemmatizer.lemmatize(x) for x in words]      # Lemmatization. 
    df['Review'][i] = ' '.join(words)                        # Rejoin the words as a sentence.
```

```python
print(df['Review'])
```

```python
#Tokenize
df['Review'] = df['Review'].apply(nltk.word_tokenize)
df['Review']
```

```python
# Removing stopwords
stop = stopwords.words('english')
df['Review'] = df['Review'].apply(lambda x: [word for word in x if word not in stop])
df['Review']
```

```python
#stemming
stemming = PorterStemmer()
df['Review'] = df['Review'].apply(lambda x: [stemming.stem(word) for word in x])
df['Review']
```
